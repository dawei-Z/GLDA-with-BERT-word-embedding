```python
import pandas as pd
import matplotlib.pyplot as plt
w2v_result = pd.read_excel("w2v_performance.xlsx")
bert_result = pd.read_excel("bert_performance.xlsx")
```

```python
bert_result.columns
```

```python
w2v_result.columns
```

```python
bert_cos = [i*10 for i in bert_result['BERT Cosine Similarity'].to_list()]
```

```python
bert_cos
```

```python
w2v_cos = w2v_result['Word2Vec Cosine Similarity']
```

```python
w2v_cos
```

```python
topics = [i for i in range(8)]
plt.title('the Mean Cosine Similarity between topics and post title')
plt.plot(topics, w2v_cos, color='green', label='Word2Vec')
plt.plot(topics, bert_cos, color='blue', label='Bert')
plt.legend()

plt.xlabel('Topic Sampler')
plt.ylabel('Cosine Similarity')
plt.show()
```

```python
bert_stab = [i*10 for i in bert_result['BERT Stability'].to_list()]
w2v_stab = [i*10 for i in w2v_result['Word2Vec Stability'].to_list()]
```

```python
topics = [i for i in range(8)]
plt.title('the Stability of each sampler')
plt.plot(topics, w2v_stab, color='green', label='Word2Vec')
plt.plot(topics, bert_stab, color='blue', label='Bert')
plt.legend()

plt.xlabel('Topic Sampler')
plt.ylabel('Stability')
plt.show()
```
